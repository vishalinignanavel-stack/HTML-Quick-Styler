```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "ibm-granite/granite-3.3-2b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)
```

```python
class ColorPaletteGenerator:
    @staticmethod
    def generate_palette(style="modern"):
        palettes = {
            "modern": {
                "primary": "#6366f1",
                "secondary": "#8b5cf6",
                "background": "#0f172a",
                "text": "#ffffff"
            },
            "ocean": {
                "primary": "#0ea5e9",
                "secondary": "#0284c7",
                "background": "#0c4a6e",
                "text": "#ffffff"
            }
        }
        return palettes.get(style.lower(), palettes["modern"])
```

```python
import gradio as gr

def generate_page(title, description, style):
    return generate_html(title, description, style)

with gr.Blocks() as demo:
    gr.Markdown("# HTML Quick Styler")

    title = gr.Textbox(label="Page Title")
    description = gr.Textbox(label="Page Description")
    style = gr.Dropdown(["modern", "ocean"], label="Color Style")

    output = gr.Code(label="Generated HTML")

    btn = gr.Button("Generate Page")
    btn.click(generate_page, inputs=[title, description, style], outputs=output)

demo.launch(share=True)
```
